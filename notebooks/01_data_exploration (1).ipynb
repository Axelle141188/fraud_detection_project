```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from datetime import datetime

# Configuration
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Blues_r")
%matplotlib inline

# Load data with correct delimiter
df = pd.read_csv(r"C:\Users\user\Desktop\fraud_detection_project\data\raw\creditcard.csv", delimiter=';')

# Display first rows
print("Data overview:")
display(df.head())

# Basic information
print("\nDataset information:")
display(df.info())

# Descriptive statistics
print("\nDescriptive statistics:")
display(df.describe())

# Check for missing values
print("\nMissing values per column:")
display(df.isnull().sum())

# Target variable distribution
fraud_count = len(df[df['Class'] == 1])
non_fraud_count = len(df[df['Class'] == 0])
print(f"\nClass distribution:")
print(f"Legitimate transactions: {non_fraud_count} ({non_fraud_count/len(df)*100:.2f}%)")
print(f"Fraudulent transactions: {fraud_count} ({fraud_count/len(df)*100:.2f}%)")
```

```python
# Import required for directory creation
import os

# Define the absolute path to your visualization directory
save_dir = r'C:\Users\user\Desktop\fraud_detection_project\visualizations\exploratory'

# Create the directory if it doesn't exist
os.makedirs(save_dir, exist_ok=True)

# 1. Transaction amount distribution
plt.figure(figsize=(10, 6))
plt.title('Transaction Amount Distribution')
sns.histplot(data=df, x='Amount', log_scale=True, hue='Class', bins=50)
plt.xlabel('Amount (log scale)')
plt.ylabel('Number of transactions')
plt.savefig(os.path.join(save_dir, 'amount_distribution.png'))
plt.show()
```

```python
# 2. Time distribution
plt.figure(figsize=(12, 6))
plt.title('Time Distribution of Transactions')
sns.histplot(data=df, x='Time', hue='Class', bins=50)
plt.xlabel('Time (seconds from start)')
plt.ylabel('Number of transactions')
plt.savefig(os.path.join(save_dir, 'time_distribution.png'))
plt.show()
```

```python
# 3. Convert Time to hours
df['Hour'] = df['Time'].apply(lambda x: (x // 3600) % 24)

plt.figure(figsize=(14, 7))
plt.title('Transaction Distribution by Hour')
# Count number of frauds by hour
fraud_by_hour = df[df['Class'] == 1].groupby('Hour').size()
normal_by_hour = df[df['Class'] == 0].groupby('Hour').size()
fraud_rate = fraud_by_hour / (fraud_by_hour + normal_by_hour)

ax = sns.barplot(x=fraud_rate.index, y=fraud_rate.values)
plt.xlabel('Hour of day')
plt.ylabel('Fraud rate')
plt.xticks(range(0, 24))
plt.savefig(os.path.join(save_dir, 'fraud_rate_by_hour.png'))
plt.show()
```

```python
# 4. Correlation heatmap
plt.figure(figsize=(16, 14))
plt.title('Correlation Matrix')
correlation = df.corr()
mask = np.triu(correlation)
sns.heatmap(correlation, annot=False, mask=mask, cmap='coolwarm', linewidths=0.5)
plt.savefig(os.path.join(save_dir, 'correlation_matrix.png'))
plt.show()
```

```python
# 5. Distribution comparison - Anonymized features
fig, axes = plt.subplots(7, 4, figsize=(20, 24))
axes = axes.flatten()

for i, feature in enumerate(df.columns[1:29]):  # Skip Time and Class
    if i < 28:  # We have 28 features
        sns.boxplot(x='Class', y=feature, data=df, ax=axes[i])
        axes[i].set_title(f'Distribution of {feature}')
        axes[i].set_xlabel('Fraud (1) vs Normal (0)')

plt.tight_layout()
plt.savefig(os.path.join(save_dir, 'feature_distributions.png'))
plt.show()
```

```python
# 6. PCA Visualization with Matplotlib (plus fiable)
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

# Data preparation
X = df.drop(['Class', 'Time'], axis=1)
y = df['Class']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

# Create dataframe 
pca_df = pd.DataFrame(
    data=X_pca, 
    columns=['PC1', 'PC2', 'PC3']
)
pca_df['Class'] = y
pca_df['Amount'] = df['Amount']

# Create 3D graph with Matplotlib
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Separate data by class
fraud = pca_df[pca_df['Class'] == 1]
normal = pca_df[pca_df['Class'] == 0]

# Plot points (sample to avoid overloading)
sample_size_normal = min(1000, len(normal))  # Limit to 1000 points for performance
normal_sample = normal.sample(sample_size_normal, random_state=42)

# Adjust point sizes for visibility
ax.scatter(normal_sample['PC1'], normal_sample['PC2'], normal_sample['PC3'], 
          c='blue', s=5, alpha=0.3, label='Normal')
ax.scatter(fraud['PC1'], fraud['PC2'], fraud['PC3'], 
          c='red', s=20, alpha=0.7, label='Fraud')

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('3D PCA Visualization of Transactions')
ax.legend()

plt.savefig(os.path.join(save_dir, 'pca_3d_matplotlib.png'))
plt.show()

# Add 2D visualization of the first two principal components
plt.figure(figsize=(12, 10))
plt.scatter(normal_sample['PC1'], normal_sample['PC2'], 
           c='blue', s=5, alpha=0.3, label='Normal')
plt.scatter(fraud['PC1'], fraud['PC2'], 
           c='red', s=20, alpha=0.7, label='Fraud')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('2D PCA Visualization (PC1 vs PC2)')
plt.legend()
plt.grid(alpha=0.3)
plt.savefig(os.path.join(save_dir, 'pca_2d_matplotlib.png'))
plt.show()
```
